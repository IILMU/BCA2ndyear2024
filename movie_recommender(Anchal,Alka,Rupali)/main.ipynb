```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
def convertcast(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
def convert_crew(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
# Give genres and keywords more weight by repeating them
def build_weighted_tags(row):
    overview  = row['overview'] if isinstance(row['overview'], list) else []
    genres    = row['genres']   if isinstance(row['genres'], list)   else []
    keywords  = row['keywords'] if isinstance(row['keywords'], list) else []
    cast      = row['cast']     if isinstance(row['cast'], list)     else []
    crew      = row['crew']     if isinstance(row['crew'], list)     else []

    # Add title words so searching "ironman" finds Iron Man
    title_words = str(row['title']).lower().replace('-', ' ').split()

    weighted = (
        title_words * 5 +   # ← title gets highest weight
        genres      * 4 +
        keywords    * 3 +
        cast        * 2 +
        crew        * 2 +
        overview    * 1
    )
    return " ".join(weighted).lower()

# Apply BEFORE stemming — replace the old tags
movies['weighted_tags'] = movies.apply(build_weighted_tags, axis=1)
new_df = movies[['movie_id', 'title', 'weighted_tags']].copy()
new_df.rename(columns={'weighted_tags': 'tags'}, inplace=True)

print("Sample weighted tag:")
print(new_df['tags'][0][:300])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#from sklearn.feature_extraction.text import TfidfVectorizer
#tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
#vectors = tfidf.fit_transform(new_df['tags'])

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf   = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(new_df['tags'])
print("Done. Vocab size:", len(tfidf.get_feature_names_out()))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def search_movies(query, top_n=10):
    # Stem the query the same way tags were stemmed
    query = query.replace('-', ' ').replace('_', ' ')
    stemmed_query = " ".join([ps.stem(w) for w in query.lower().split()])
    
    # Vectorize query using the SAME fitted tfidf
    query_vec = tfidf.transform([stemmed_query])
    
    # Compute similarity between query and all movies
    scores = cosine_similarity(query_vec, vectors).flatten()
    
    # Get top N movie indices
    top_indices = np.argsort(scores)[::-1][:top_n]
    
    results = []
    for idx in top_indices:
        if scores[idx] > 0:  # only return actual matches
            results.append({
                'title': new_df.iloc[idx]['title'],
                'score': round(float(scores[idx]), 4)
            })
    return results
```

```python
new_df = new_df.merge(ratings, on='title', how='left')

def search_movies_ranked(query, top_n=10):
    query = query.replace('-', ' ').replace('_', ' ')
    stemmed_query = " ".join([ps.stem(w) for w in query.lower().split()])
    query_vec = tfidf.transform([stemmed_query])
    scores = cosine_similarity(query_vec, vectors).flatten()
    top_indices = np.argsort(scores)[::-1][:50]  # get top 50, re-rank

    results = []
    for idx in top_indices:
        if scores[idx] > 0:
            row = new_df.iloc[idx]
            # 70% keyword match + 30% normalized rating
            combined = 0.7 * scores[idx] + 0.3 * (row['vote_average'] / 10)
            results.append({
                'title': row['title'],
                'similarity': round(float(scores[idx]), 4),
                'rating': row['vote_average'],
                'score': round(combined, 4)
            })

    results.sort(key=lambda x: x['score'], reverse=True)
    return results[:top_n]
```
